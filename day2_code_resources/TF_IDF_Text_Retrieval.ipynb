```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# vec1, vec2
# cos(vec1, vec2) = tích vô hướng vec1, vec2 / tích độ dài |vec1|, |vec2|
```

```python
vi_data = pd.read_csv("/content/drive/MyDrive/ML_DL_datasets/vi_text_retrieval.csv")
vi_data.info()
```

```python
vi_data.head()
```

```python
texts = vi_data["text"]
lower_texts = [text.lower() for text in texts]
```

```python
# Khởi tạo TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
text_embedded = tfidf_vectorizer.fit_transform(lower_texts).toarray()
```

```python
text_embedded[7]
```

```python
# Hàm tf_idf search
def corr_search(question, tfidf_vectorizer, text_embedded, top_d=5):
    # Lower question
    question = question.lower()
    # Vector hoá câu hỏi
    query_embedded = tfidf_vectorizer.transform([question]).toarray()[0]
    # Tính cosine similarity score
    cosine_scores = cosine_similarity([query_embedded], text_embedded)[0]

    # Lấy ra top document có điểm tương đồng đối với câu hỏi cao nhất
    results = []
    for idx in cosine_scores.argsort()[-top_d:][::-1]:
        text_score = {
            'id': idx,
            'corr_score': cosine_scores[idx]
        }
        results.append(text_score)
    return results
```

```python
question = "Quân đội Hoa Kỳ gồm những lực lượng nào"
results = corr_search(question, tfidf_vectorizer, text_embedded)
results
```

```python
for result in results:
    answer = vi_data.loc[result['id']].text
    print(answer)
    print()
```
